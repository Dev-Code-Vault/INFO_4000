### Replay Buffer ###

```python
import gymnasium as gym
from collections import deque
import random
```

```python
env = gym.make('CartPole-v1')
```

```python
# Generating individual experiences

def replay_buffer(action):
    currentState = env.reset()[0].tolist()
    obs,reward,done,_,_ = env.step(action)
    experience  = currentState,action,reward,obs.tolist(),done
    return experience
```

```python
# Creating the replay buffer with experiences

replay_buffermax = 10000
replay = deque(maxlen=replay_buffermax)  # deque is a Python function which pops out the oldest item and adds the latest

for i in range(replay_buffermax):
    a = random.randint(0,1)
    e = replay_buffer(a)
    replay.append(e)
```

```python
print(f"Total rows: {len(replay)}\n\nEach row in the buffer is a tuple of (s,a,r,s'):\n\n{replay[-2]}")
```
