### Finding Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
# Example vectors
vec_a = np.array([[1, 2, 3]])
vec_b = np.array([[2, 3, 4]])
vec_c = np.array([[100, 10, -10]])

# Compute cosine similarity
similarity_ab = cosine_similarity(vec_a, vec_b)
similarity_ac = cosine_similarity(vec_a, vec_c)

print("Cosine similarity between A and B:", similarity_ab[0][0])
print("Cosine similarity between A and C:", similarity_ac[0][0])

# Example with multiple vectors
vectors = np.array([
    [1, 2, 3],
    [2, 3, 4],
    [100, 10, -10]
])

similarity_matrix = cosine_similarity(vectors)

print("\nCosine similarity matrix:")
print(similarity_matrix)
```
