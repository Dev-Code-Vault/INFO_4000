#### Using the Pipeline API for text generation
Hugging Face's pipeline API simplifies the process by providing high-level abstractions for various tasks.

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
from hf_access import ACCESS_TOKEN
from huggingface_hub import login

hfaccess = ACCESS_TOKEN


login(token=hfaccess)

model = AutoModelForCausalLM.from_pretrained(
            "google/gemma-3-270m",#"mlfoundations-dev/oh-dcft-v3.1-gemini-1.5-flash", 
            device_map='cpu', 
            trust_remote_code=True,
            torch_dtype="auto"
    )
tokenizer = AutoTokenizer.from_pretrained("google/gemma-3-270m", trust_remote_code=True)
        
# Create a text generation pipeline
pipe = pipeline(
    "text-generation", 
    model=model, 
    tokenizer=tokenizer,
    max_new_tokens=50,
    do_sample=True,
    temperature=0.7,
    top_k=50,
    top_p=0.95
) 
```

```python
result = pipe("Physics is")
print(result[0]['generated_text'])
```
