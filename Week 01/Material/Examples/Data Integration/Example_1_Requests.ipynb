### Using 'requests' module to access APIs ###

```python
import requests
import json
```

```python
# People in space on ISS

url_astro = "http://api.open-notify.org/astros.json"

astros = requests.request("GET",url_astro)
print(f"Response Status Code:{astros}\n")

result2 = json.loads(astros.text) # Assign content of response to a variable
print(f" Response Text:\n{result2}\n")
print(f"Dict keys in response:\n{result2.keys()}\n")
print(f"\nNumber of astros on ISS\n{result2['number']}\n")

who = result2['people']

print(f"\nAstros on board ISS\n")
for i in who:
    print(i['name'])
```

```python
# Accessing another NASA API - p0sition of ISS

url_nasa = "http://api.open-notify.org/iss-now.json"

position = requests.request("GET",url_nasa)
print(f"Status Code:{position}\n")
print(position.headers)
result1 = json.loads(position.text)
print(f"\nKeys:\n{result1.keys()}")
print(result1)
```
